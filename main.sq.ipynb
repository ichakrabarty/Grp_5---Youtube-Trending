```python
#importing dependencies
import pandas as pd
from datetime import datetime as dt
```

```python
df= pd.read_csv('Datasources/US_youtube_trending_data.csv')
df
```

```python
#leaving the columns we want to work with 
youtube_df= df[['video_id','title','publishedAt','channelId','channelTitle','categoryId','trending_date', 'tags','view_count','likes','dislikes' ]]

youtube_df
```

```python
youtube_df['publishedAt'].astype
```

```python
#transforming the datetime to new format
youtube_df['publishedAt'] = youtube_df['publishedAt'].str.split('T').str[0]
youtube_df['trending_date'] = youtube_df['trending_date'].str.split('T').str[0]
youtube_df.head()
```

```python
dict = {'video_id': 'Video_id',
        'title': 'Title',
        'publishedAt': 'Published_date',
        'channelId': 'Chanel_id',
        'channelTitle': 'Channel_Title',
        'categoryId':'Category_id',
        'trending_date':"Trending_date" ,
        'tags':'Tags', 
        'view_count':'View_count', 
        'likes':'Likes',
        'dislikes':'Dislikes'            
                
        
       }
  
# call rename () method
youtube_df.rename(columns=dict,
          inplace=True)
  
# print Data frame after rename columns
youtube_df
```

```python
youtube_df.drop('Tags', axis=1, inplace=True)
youtube_df
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/youtube_db')
connection = engine.connect()
```

```python
engine.table_names()
```

```python
categories_df.to_sql(name='category', con=engine, if_exists='append', index=False)
```

```python
channel_df.to_sql(name='channel', con=engine, if_exists='append', index=False)
```

```python
title_df.to_sql(name='titles', con=engine, if_exists='append', index=False)
```
